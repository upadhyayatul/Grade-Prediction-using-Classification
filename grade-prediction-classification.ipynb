### Data Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df.Grade = le.fit_transform(df.Grade)
df.ScholarType = le.fit_transform(df.ScholarType)
df.Direction = le.fit_transform(df.Direction)
df.Gender = le.fit_transform(df.Gender)
df.Medium = le.fit_transform(df.Medium)
df.CourseType = le.fit_transform(df.CourseType)
df.ProgramType = le.fit_transform(df.ProgramType)
```

```python
from sklearn.preprocessing import Imputer
imr = Imputer(missing_values = 'NaN', strategy = 'mean')
imr = imr.fit(df.iloc[:,3:8])
df.iloc[:,3:8] = imr.transform(df.iloc[:,3:8])
```

### Data Training and Testing

```python
from sklearn.model_selection import train_test_split
x = df.iloc[:,3:8]
y = df.iloc[:,2]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

### Feature Scaling

#### Standardisation

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train_ = ss.fit_transform(x_train)
x_test_ = ss.transform(x_test)
```

#### Normalization

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
x_train_ = mms.fit_transform(x_train)
x_test_ = mms.transform(x_test)
```

```python
x_train_
```

```python
print(x_train_.shape)
print(x_train.shape)
```

```python
x_test_
```

### Classification

#### GaussianNB

```python
accuracy = []
f1 = []
precision = []
recall = []
classifiers = ['GaussianNB','SVM','Bagging','Boosting','Decision Tree','KNN','Logistic Regression']
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train_,y_train)
y_test_ = gnb.predict(x_test_)
```

```python
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score,classification_report
print(accuracy_score(y_test,y_test_))
print(f1_score(y_test,y_test_,average = 'weighted'))
print(precision_score(y_test,y_test_,average = 'macro'))
print(recall_score(y_test,y_test_,average = 'macro'))
print(classification_report(y_test,y_test_))
accuracy.append(accuracy_score(y_test,y_test_))
f1.append(f1_score(y_test,y_test_,average = 'weighted'))
precision.append(precision_score(y_test,y_test_,average = 'macro'))
recall.append(recall_score(y_test,y_test_,average = 'macro'))
```

```python
accuracy
```
